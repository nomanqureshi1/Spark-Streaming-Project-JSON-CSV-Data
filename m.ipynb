```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder \
    .appName("Read CSV Files") \
    .getOrCreate()

# Read all CSV files from the folder
df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("/Users/nomantahir/Desktop/Project-Spark_streaming/output/*.csv")

# Show the data
df.show(100)
```
